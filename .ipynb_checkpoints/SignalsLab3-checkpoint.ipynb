```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
import sys


def visualize_region(timestamps, signal, t0, t1):
    if t0 >= t1:
        raise ValueError("t1 cannot be lower than t0")
    current_timestamps_indexes = np.where((timestamps > t0) &(timestamps < t1) )[0]
    first_index, last_index = max(0,current_timestamps_indexes[0] - 1), current_timestamps_indexes[-1] + 2
    current_timestamps = timestamps[first_index:last_index]
    current_signal = signal[first_index:last_index]

    return current_timestamps, current_signal

def signal_plot(timestamps, signal_range, t1=None, t2=None, x_label=None, y_label=None, title=None, extraticks=None, function=plt.plot):
    # plt.rcParams["figure.figsize"] = figsize
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    if t1 is not None and t2 is not None:
        out = visualize_region(timestamps, signal_range, t1, t2)
        function(out[0], out[1])
    else:
        function(timestamps, signal_range)

def return_timestamps(fs, duration):
    return np.linspace(0, duration, int(fs*duration))

def sin_signal(fs, sin_timestamps, amplitude):
    return amplitude * np.sin(2*np.pi*sin_timestamps*fs)

def form_sin_signal(descr_fs, sin_fs, duration, amplitude):
    timestmps = return_timestamps(descr_fs, duration)
    return timestmps, sin_signal(sin_fs, timestmps, amplitude)

def fft(fs, signals):
    ft = np.fft.fft(signals)
    Aspectrum = 2 * np.abs(ft)[:fs] / len(signals)
    Pspectrum = np.angle(ft) * (180 / np.pi)
    return Aspectrum, Pspectrum
```
